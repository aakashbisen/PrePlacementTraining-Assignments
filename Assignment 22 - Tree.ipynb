'''<aside>
ðŸ’¡ Question-1:

Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.

Example:

![Untitled](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/f9dda6ae-60b9-427b-990f-dc5b3117a5e3/Untitled.png)


```python
#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def convertToDLL(root):
    if root is None:
        return None

    # Convert the left subtree
    left_head = convertToDLL(root.left)

    # Find the rightmost node in the left subtree
    rightmost = left_head
    if rightmost:
        while rightmost.right:
            rightmost = rightmost.right

    # Convert the current node
    if left_head:
        rightmost.right = root
        root.left = rightmost
    else:
        left_head = root

    # Convert the right subtree
    right_head = convertToDLL(root.right)

    # Connect the right subtree to the current node
    if right_head:
        right_head.left = root
        root.right = right_head

    return left_head

# Test case
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

result = convertToDLL(root)

# Print the doubly linked list in the order of an inorder traversal
def printDLLInOrder(head):
    node = head
    while node:
        print(node.val, end=" ")
        node = node.right

printDLLInOrder(result)  # Output: 5 10 30 20 35
```

```python
'''<aside>
ðŸ’¡ Question-2

A Given a binary tree, the task is to flip the binary tree towards the right direction that is clockwise. See the below examples to see the transformation.

In the flip operation, the leftmost node becomes the root of the flipped tree and its parent becomes its right child and the right sibling becomes its left child and the same should be done for all left most nodes recursively.

</aside>'''

#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def flipBinaryTree(root):
    if root is None:
        return None

    # Base case: leaf node
    if root.left is None and root.right is None:
        return root

    # Flip the left and right subtrees recursively
    flipped_left = flipBinaryTree(root.left)
    flipped_right = flipBinaryTree(root.right)

    # Update the children of the current node
    root.left = flipped_right
    root.right = flipped_left

    return root

# Test case
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

flipped_tree = flipBinaryTree(root)

# Print the flipped tree
def printTree(root):
    if root is None:
        return

    print(root.val, end=" ")
    printTree(root.left)
    printTree(root.right)

printTree(flipped_tree)  # Output: 1 3 2 5 4
```

```python
'''<aside>
ðŸ’¡ Question-3:

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

Input:

        6
     /    \
    3      5
  /   \     \
 2     5     4
     /   \
    7     4

Output:

There are 4 leaves, hence 4 root to leaf paths -
  6->3->2
  6->3->5->7
  6->3->5->4
  6->5>4

</aside>'''


#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def printRootToLeafPaths(root):
    if root is None:
        return

    stack = [(root, str(root.val))]

    while stack:
        node, path = stack.pop()

        if node.left is None and node.right is None:
            print(path)
        
        if node.right:
            stack.append((node.right, path + "->" + str(node.right.val)))
        if node.left:
            stack.append((node.left, path + "->" + str(node.left.val)))

# Test case
root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(5)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)
root.right.right = TreeNode(4)

printRootToLeafPaths(root)
```

```python
'''
ðŸ’¡ Question-4:

GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

**Examples:**

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 2 4 5 3
        Postorder -> 4 5 2 3 1
Output : 

Yes
Explanation : 

All of the above three traversals are of
the same tree 

                           1
                         /   \
                        2     3
                      /   \
                     4     5

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 5 4 2 3
        Postorder -> 4 1 2 3 5
Output : 

No
'''

#Answer:
def isSameTree(inorder, preorder, postorder):
    # Base case: empty tree
    if not inorder:
        return True

    # Check if the root values are the same
    if preorder[0] != postorder[-1]:
        return False

    # Find the root value in the inorder traversal
    root_val = preorder[0]
    root_index = inorder.index(root_val)

    # Recursively check the left and right subtrees
    left_inorder = inorder[:root_index]
    right_inorder = inorder[root_index + 1:]

    left_preorder = preorder[1:root_index + 1]
    right_preorder = preorder[root_index + 1:]

    left_postorder = postorder[:root_index]
    right_postorder = postorder[root_index:-1]

    return (
        isSameTree(left_inorder, left_preorder, left_postorder) and
        isSameTree(right_inorder, right_preorder, right_postorder)
    )

# Test cases
inorder = [4, 2, 5, 1, 3]
preorder = [1, 2, 4, 5, 3]
postorder = [4, 5, 2, 3, 1]
print(isSameTree(inorder, preorder, postorder))  # Output: True

inorder = [4, 2, 5, 1, 3]
preorder = [1, 5, 4, 2, 3]
postorder = [4, 1, 2, 3, 5]
print(isSameTree(inorder, preorder, postorder))  # Output: False
```
