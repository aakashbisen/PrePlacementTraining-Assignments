```python
'''<aside>
ðŸ’¡ Question-1

You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

</aside>'''


#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def convertToBST(root):
    # Step 1: Perform in-order traversal and extract values
    values = []
    inorderTraversal(root, values)

    # Step 2: Sort the extracted values
    values.sort()

    # Step 3: Perform in-order traversal and assign sorted values
    index = [0]  # Wrapper to keep track of current index in values
    assignValues(root, values, index)

    return root

def inorderTraversal(node, values):
    if node is None:
        return

    inorderTraversal(node.left, values)
    values.append(node.val)
    inorderTraversal(node.right, values)

def assignValues(node, values, index):
    if node is None:
        return

    assignValues(node.left, values, index)
    node.val = values[index[0]]
    index[0] += 1
    assignValues(node.right, values, index)

# Test case
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

result = convertToBST(root)

# Print the converted binary search tree (in-order traversal)
def inorderTraversalPrint(node):
    if node is None:
        return

    inorderTraversalPrint(node.left)
    print(node.val, end=" ")
    inorderTraversalPrint(node.right)

inorderTraversalPrint(result)  # Output: 2 4 7 8 10
```

```python
'''<aside>
    
ðŸ’¡ Question-2:

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

Consider the following BST:

</aside>
**Input-1:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 6

node-2 = 14

**Output-1:**

The distance between the two keys = 4

**Input-2:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 3

node-2 = 4

**Output-2:**

The distance between the two keys = 2'''


#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def findDistance(root, node1, node2):
    lca = findLCA(root, node1, node2)
    distance1 = findDistanceToNode(lca, node1)
    distance2 = findDistanceToNode(lca, node2)
    return distance1 + distance2

def findLCA(root, node1, node2):
    if root.val > node1.val and root.val > node2.val:
        return findLCA(root.left, node1, node2)
    elif root.val < node1.val and root.val < node2.val:
        return findLCA(root.right, node1, node2)
    else:
        return root

def findDistanceToNode(root, node):
    if root.val == node.val:
        return 0
    elif root.val > node.val:
        return 1 + findDistanceToNode(root.left, node)
    else:
        return 1 + findDistanceToNode(root.right, node)

# Test case
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right = TreeNode(14)
root.right.right.left = TreeNode(13)

node1 = TreeNode(6)
node2 = TreeNode(14)

distance = findDistance(root, node1, node2)
print("The distance between the two keys =", distance)  # Output: 4
```

```python
'''<aside>
ðŸ’¡ Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

</aside>'''


#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class DoublyLinkedListNode:
    def __init__(self, val=0, prev=None, next=None):
        self.val = val
        self.prev = prev
        self.next = next

def convertToDoublyLinkedList(root):
    if root is None:
        return None

    # Perform in-order traversal
    dummy = DoublyLinkedListNode()
    prev = dummy
    stack = []
    node = root

    while node or stack:
        while node:
            stack.append(node)
            node = node.left

        node = stack.pop()
        curr = DoublyLinkedListNode(node.val)
        prev.next = curr
        curr.prev = prev
        prev = curr

        node = node.right

    head = dummy.next
    if head:
        head.prev = None

    return head

# Test case
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

result = convertToDoublyLinkedList(root)

# Print the doubly linked list
def printDoublyLinkedList(head):
    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.next

printDoublyLinkedList(result)  # Output: 5 10 30 20 35
```

```python
'''<aside>
ðŸ’¡ Question-4:

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 â†’ -1

2 â†’ 3

3 â†’ -1

4 â†’ 5

5 â†’ 6

6 â†’ 7

7 â†’ -1

</aside>'''


#Answer:
class TreeNode:
    def __init__(self, val=0, left=None, right=None, next=None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next

def connectNodesAtSameLevel(root):
    if root is None:
        return

    # Perform level order traversal
    queue = [root]

    while queue:
        level_size = len(queue)

        for i in range(level_size):
            node = queue.pop(0)

            if i < level_size - 1:
                node.next = queue[0]

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return root

# Test case
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

result = connectNodesAtSameLevel(root)

# Print the connections between nodes
def printNodeConnections(root):
    node = root
    while node:
        curr = node
        while curr:
            if curr.next:
                print(curr.val, "â†’", curr.next.val)
            else:
                print(curr.val, "â†’ -1")
            curr = curr.next
        node = node.left

printNodeConnections(result)
```
